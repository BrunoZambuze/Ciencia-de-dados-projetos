## Visualiza√ß√£o dos Dados

```python
sns.pairplot(df_cancer, hue = 'target', vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness']);
```

## Treino Modelo

```python
X = df_cancer.drop(['target'], axis = 1)
```

```python
y = df_cancer['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train, y_train)
```

## Avalia√ß√£o do Modelo

```python
y_predict = svc_model.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot = True);
```

```python
accuracy_score(y_test, y_predict)
```

```python
print(classification_report(y_test, y_predict))
```

# Melhorando o Modelo

### Normaliza√ß√£o das features

```python
min_train = X_train.min()
```

```python
range_train = (X_train - min_train).max()
```

```python
X_train_scaled = (X_train - min_train) / range_train
```

```python
sns.scatterplot(x = X_train['mean area'], y = X_train['mean smoothness'], hue = y_train);
#Valores antes do escalonamento
```

```python
sns.scatterplot(x = X_train_scaled['mean area'], y = X_train_scaled['mean smoothness'], hue = y_train);
#Valores ap√≥s o escalonamento
```

```python
min_test = X_test.min()
```

```python
range_test = (X_test - min_test).max()
```

```python
X_test_scaled = (X_test - min_test) / range_test
```

```python
svc_model.fit(X_train_scaled, y_train)
```

```python
y_predict = svc_model.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot = True);
```

```python
accuracy_score(y_test, y_predict)
```

```python
print(classification_report(y_test, y_predict))
```

### Tuning dos par√¢metros com GridSearch

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {
    'tol': [0.001, 0.0001, 0.00001, 0.000001],
    'C': [0.1, 1, 10, 1000],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'sigmoid']
}
```

```python
grid_search = GridSearchCV(SVC(), param_grid = parameters, refit = True, verbose = 4)
```

```python
grid_search.fit(X_train_scaled, y_train)
```

```python
print(grid_search.best_params_)
```

```python
print(grid_search.best_score_)
```

```python
grid_predict = grid_search.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, grid_predict)
```

```python
sns.heatmap(cm, annot = True);
```

```python
accuracy_score(y_test, grid_predict)
```

```python
print(classification_report(y_test, grid_predict))
```
